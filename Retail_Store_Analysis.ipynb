```python
df = pd.merge(Order, data, on="Order ID")
```

```python
df.shape
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month_name()
df['Year'] = df['Order Date'].dt.year
```

```python
total_sales = df['Amount'].sum()
print(total_sales)
```

```python
total_sales = df['Profit'].sum()
print(total_sales)
```

```python
total_sales = df['Quantity'].sum()
print(total_sales)
```

```python
category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(category_sales)
```

```python
category_sales.plot(kind='bar',color='red')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
category_profit = df.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar')
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```

```python
monthly_sales = df.groupby('Month')['Amount'].sum()
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
top_states = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10)
top_states.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.show()
```

```python
top_customers = df.groupby('CustomerName')['Amount'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
top_customers.plot(kind='bar')
```

```python
plt.scatter(df['Amount'], df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
numeric_df = df[['Amount', 'Profit', 'Quantity', 'AOV']]
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
